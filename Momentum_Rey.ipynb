```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np
import math
import warnings
from scipy.stats import norm
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
from datetime import date,datetime
from pandas.tseries.offsets import MonthEnd
warnings.filterwarnings('ignore')
```

```python
equity=pd.read_excel('equity.xlsx')
equity['Date'] = pd.to_datetime(equity['Date'])
equity.set_index('Date',inplace=True)
equity = equity.sort_index(ascending=True)
equity = equity.loc[(equity.index>"2000-12-30") & (equity.index<"2020-09-01")]
equity = equity.loc[equity.reset_index().groupby(equity.index.to_period('M'))['Date'].max()]
equity.head(15)
```

```python
Equity12monthReturn = equity.pct_change(12)
Equity12monthReturn.head(15)
```

```python
Equity12monthReturn = Equity12monthReturn.shift()
Equity12monthReturn.head(15)
```

```python
Equity12monthReturn[Equity12monthReturn > 0] = 1
Equity12monthReturn[Equity12monthReturn < 0] = -1
Equity12monthReturn.head(15)
```

```python
#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)
#pd.set_option('display.width', None)
#pd.set_option('display.max_colwidth', -1)
EquityMonthlyReturn = equity.pct_change()
EquityMonthlyReturn.head(15)
```

```python
Momentum = Equity12monthReturn*EquityMonthlyReturn
Momentum = Momentum.loc[(Momentum.index> "2000-12-31")]
Momentum += 1
Momentum.head(20)
```

```python
Momentum = Momentum.loc[(Momentum.index>"2001-12-31")]
Momentum
```

```python
TotalReturn = Momentum.prod(axis = 0)
TotalReturn = TotalReturn.to_frame()
TotalReturn = TotalReturn.transpose()
AnnualReturn = TotalReturn.pow(12/224) - 1 
AnnualReturn.index = ['Annual Return']
AnnualReturn
```

```python
MomentumAnnual = Momentum.groupby(pd.Grouper(freq="Y")).prod()
MomentumAnnual
```

```python
MomentumAnnual -= 1
MomentumAnnual.index = [2002, 2003, 2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
MomentumAnnual.head(15)
```

```python
MomentumStrategy = pd.concat([MomentumAnnual,AnnualReturn])
MomentumStrategy
```

```python
MomentumAnnualdata = MomentumAnnual.describe()
MomentumAnnualdata = MomentumAnnualdata.drop(MomentumAnnualdata.index[0])
MomentumStrategy = pd.concat([MomentumAnnual,MomentumAnnualdata])
MomentumStrategy = MomentumStrategy[:-5]
MomentumStrategy
```

```python
MomentumStrategy.index = [2002, 2003, 2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020, 'Annual Return', 'Annual Vol']
MomentumStrategy
```

```python
MomentumStrategy.loc['Sharpe Ratio']= MomentumStrategy.loc['Annual Return']/MomentumStrategy.loc['Annual Vol']
```
